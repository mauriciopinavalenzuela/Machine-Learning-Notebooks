# Entendiendo los datos

## Cargar los datos desde una url pública.

Hay varias formas de alimentar los datos en colab, pero en este primer ejemplo voy a cargar los datos desde una url pública.
Yo NUNCA cargo, subo los datos directamente a colab; prefiero dejarlos en algún lugar previamente (en drive o en un servidor público, como github).

Para cargar los datos usamos la biblioteca Pandas.


```python
import pandas as pd

url_base = 'https://dmontero.org/mineria/datasets/'
url = url_base + 'Food_Preference.csv'
df = pd.read_csv(url)
```

## Miramos unos pocos de los datos cargado
No queremos mirar todos los datos, porque suelen ser demasiados!!
El archivo venía en formato csv (valores separados por comas) uno de los formatos comunes para los datos. Otros dos formatos son frecuentes: json y parquet.
La estructura de datos para datos tabulares de pandas se llama dataframe, abreviado 'df'.

```python
df.head(10)
```

Ahí vemos los primeros datos.

Pero a veces quiere ver una cantidad de datos que no sean los del principio.

```python
df.sample(10)
```

¿Cuántos datos hay?

```python
df.shape
```

Puedo calcular algunos estadísticos para tratar de entender que pasa.

```python
df['Age'].mean()
```

O tal vez un poco más ordenado

```python
print('Promedio: ' , df['Age'].mean())
print('Desviación estándar: ' , df['Age'].std())
print('Mediana: ' , df['Age'].median())
```

## Gráficar los datos.

Para eso debemos usar una biblioteca de gráficos, en este caso pyplot.

```python
import matplotlib.pyplot as plt
etiquetas = df['Food'].unique()
valores = df['Food'].value_counts()

plt.pie(valores, labels=etiquetas, autopct='%1.1f%%')
plt.title('Preferencias de alimentación')
plt.show()
```

## Interpretación

Al parecer los datos representan a una encuesta de preferencias de alimentación hecha en la India en 2019.
